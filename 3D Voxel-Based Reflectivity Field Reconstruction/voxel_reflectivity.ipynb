```python

import numpy as np
```

```python
def voxelize_points(points, intensity, voxel_size):
    """
    Assign points to voxels and build voxel-wise reflectivity statistics.
    """
    voxel_indices = np.floor(points / voxel_size).astype(int)

    voxel_dict = {}

    for idx, v in enumerate(voxel_indices):
        key = tuple(v)

        if key not in voxel_dict:
            voxel_dict[key] = {
                "intensity": [],
                "count": 0
            }

        voxel_dict[key]["intensity"].append(intensity[idx])
        voxel_dict[key]["count"] += 1

    return voxel_dict
```

```python
def compute_voxel_reflectivity(voxel_dict):
    """
    Compute voxel-level reflectivity features.
    """
    voxel_features = {}

    for key, data in voxel_dict.items():
        intensities = np.array(data["intensity"])

        voxel_features[key] = {
            "mean_intensity": np.mean(intensities),
            "var_intensity": np.var(intensities),
            "point_count": data["count"]
        }

    return voxel_features
```

```python
def local_normalization(voxel_features):
    """
    Normalize voxel reflectivity across neighboring voxels.
    """
    means = np.array([v["mean_intensity"] for v in voxel_features.values()])
    global_mean = np.mean(means)
    global_std = np.std(means) + 1e-6

    for v in voxel_features.values():
        v["normalized_reflectivity"] = (
            v["mean_intensity"] - global_mean
        ) / global_std

    return voxel_features
```
